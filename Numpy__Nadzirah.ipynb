```python
# Python List vs Numpy

# how much execution time will be taken by List and sum()
import time
numbers = list(range(1,100000001))
total = 0
start_time = time.time()  # record start time
for i in numbers:
  total = total + i
end_time = time.time()  # record end time
print("Total : ",total)
print("Execution Time : ",end_time - start_time,"seconds")
print(f"Execution Time : {end_time - start_time:.4f} seconds")
#.4f means show 4 digits after the decimal point
```

```python
numbers = np.arange(1,100000001)
total = 0
start_time = time.time()  # record start time
total = np.sum(numbers)
end_time = time.time()  # record end time
print("Total : ",total)
print("Execution Time : ",end_time - start_time,"seconds")
print(f"Execution Time : {end_time - start_time:.4f} seconds")
```

```python
# copy() vs view() methods of Numpy

# copy()
# a copy will create a new independent array in a different memory location

org_array = np.array([1,2,3,4,5,6])
print(org_array)

copied_arr = org_array.copy()
print(copied_arr)

# if we make any changes in copied array , it won't effect the original array
copied_arr[0] = 100
print(copied_arr)
print(org_array)
```

```python
#cubaan pertama copy saya
org_array = np.array([10,20,30])
copied_arr = org_array.copy()
print(copied_arr)

copied_arr[0] = 25
print(org_array)
```

```python
# view()
# a view is the representation of the original array
# once changes made, developer will get notified

org_array = np.array([1,2,3,4,5,6])
print(org_array)

copied_arr = org_array.view()
print(copied_arr)

# if we make any changes in copied array , it will effect the original array
copied_arr[0] = 100
print(copied_arr)
print(org_array)

# if we make any changes in org array , it will effect the copied array
org_array[4] = 400
print(org_array)
print(copied_arr)
```

```python
org_array = np.array([1,2,3,4,5,6])
copied_arr = org_array.view()

# if use view, both will same shares_memory() --> will return True if arrays are sharing the same memory location

# syntax to see if its true
print(np.shares_memory(org_array,copied_arr))
```

```python
# concatenate()
# used to join 2 arrays
# METHOD NOT SO IMPORTANT BUT SOMETIMES CAN BE USEFUL
# arr merupakan array
# array tu macam list tapi dalam numpy panggil arr
# concatenate method gabungkan 2 arr
# tapi cuma boleh gabung kalau dua dua ni sama DIMENSI

pp1 = np.array([20,60,80,45,32])
pp2 = np.array([10,20,30,40,50,67,32,90])
pp3 = np.array([100,200,300,400])
arr = np.concatenate((pp1,pp2,pp3))
print(arr)
```

```python
# Sorting
# np.sort() by default it will sort in an ascending order

prices = np.array([78,34,56,90,102,34,12,67,172,120,95,45])
print(np.sort(prices))
```

```python
# Filter and Searching In An Array
# dekat sini kita guna for loop
prices = np.array([78,34,56,90,102,34,12,67,172,120,95,45])
print(prices)

for i in prices:
  if i>=50:
    print(i)
```

```python
# second method untuk filter and search
# Boleh guna Boolean Masking
prices = np.array([78,34,56,90,102,34,12,67,172,120,95,45])
prices>=50
```

```python
prices = np.array([78,34,56,90,102,34,12,67,172,120,95,45])
# Boolean Masking
newarr = prices[prices>=50]  # here we are applying a filter prices[showMeOnlyTrueValues]
print(newarr)
```

```python
prices = np.array([78,34,56,90,102,34,12,67,172,120,95,45])
# show the even number prices
newarr = prices[prices%2==0]
print(newarr)
```

```python
# np.where(expression)
prices = np.array([78,34,56,90,102,34,12,67,172,120,95,45])
# np.where returns INDEX NUMBERS
res = np.where(prices%2==0)
print(prices[res])
```

```python
prices = np.array([78,34,56,90,102,34,12,67,172,120,95,45])
res = np.where((prices%2==0) & (prices>=50))
print(prices[res])

# & combines both conditions
```

```python
prices = np.array([78,34,56,90,102,34,12,67,172,120,95,45])
print(prices[(prices%2==0) & (prices>=50)])
```

```python
# SALARY CLASSIFICATION
salary = np.array([25000, 40000, 55000, 70000, 30000, 90000])
# Replace salary with High if salary >= 60000
# Replace salary with Medium if salary >= 40000 and <= 60000
# Otherwise Low

# syntax:
# np.where(condition,value_if_true,value_if_false)

result = np.where(
    salary >= 60000 , "High" ,
    np.where(salary >=40000 , "Medium" ,"Low")
)

print(result)
```

```python
# Numpy Broadcasting
arr1 = np.array([16,24,26])
arr2 = np.array([13,22,19])

print(arr1+arr2)
print(arr1*arr2)
```

```python
arr1 = np.array([16,24,26,12]) # dekat sini ada extra shape, tiba tiba ada 12
arr2 = np.array([13,22,19]) # array tak sama shape
print(arr1+arr2)
# akan dapat error sebab tak sama Dimension
# error ni dipanggil broadcasting
```

```python
# in which cases there will be no broadcasting error
# case 1 : when arrays are equal
# case 2 : if anyone of them is 1

arr1 = np.array([16,24,26])
arr2 = np.array([2])            # 2 2 2

print(arr1+arr2)
```
