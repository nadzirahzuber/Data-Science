```python
# Pandas Data Science Library
# Numpy --> Numerical Python

import numpy as np
import pandas as pd

print(pd.__version__)
```

```python
# numpy --> array()
arr = np.array([10, 20, 30, 40, 50])
print(type(arr))
print(arr)
```

```python
# Pandas Series
# Series is 1D Labelled Array 0,1,2,3,4 bahagian row
# Data frame for 2D
s = pd.Series([10, 20, 30, 40, 50])
print(type(s))
print(s)
print(s.shape)

# In Numpy --> Only Numbers
# In Series --> Numbers + Labels (index)
```

```python
info = pd.Series(['John', 28, 6700, 'Klang'], index = ['name', 'age', 'sal', 'city'])
print(info)
print(info['name'])      # indexing
print(info.get('city'))  # getting data values using get()
print(info.iloc[2])      # iloc --> index location

# Series is like a Dictionary + combo
```

```python
info = {'Name':'John', 'Age':28, 'Sal':6700, 'City':'Klang'}
print(type(info))
print(info['Name'])
print(info.get('City'))
print(info[0])
```

```python
# value property
# print only value

print(info.values)
```

```python
# Indexing Slicing and Using For Loop
s = pd.Series([100, 200, 300, 400, 500, 600], index=['v1', 'v2', 'v3', 'v4', 'v5', 'v6'])
print(s)

# 100  200  300  400  500  600
#  0    1    2    3    4    5
# -6   -5   -4   -3   -2   -1
# v1   v2   v3   v4   v5   v6

# Indexing in Series can be done by using [] or iloc[]  or loc[]

print('\nIndexing In Series')
print(s.iloc[0])     # for index numbers
print(s.loc['v5'])   # loc used for labels
print(s['v2'])       # [] used for labels
print(s.iloc[-1])


print('\nSlicing In Series')
print(s[1:4])         # 4 will be excluded --> stopIndex
print(s['v1':'v5'])   # if performing slicing using labels, the end label is included
print(s.iloc[0:4])

# print 100 , 400 and 600 only
print(s.iloc[[0, 3, 5]])
print(s[['v1','v4','v6']])  # for specific information use s[['','','']]
print(s[['v1','v4','v6']].values)

# Conditional selection
print('\nConditional Selection')
print(s[s%3==0])
print(s[s>300])
```

```python
# Using for Loop

for i in s:
  print(i)
```

```python
for k,v in s.items():
  print(k,v)
```

```python
for i in enumerate(s):
  print(i)
```
